```python
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Adquisición imagen

Para manipular de mejor manera la imágen y mejorar el tiempo de corrida. Se puede escoger 
la cantidad de filas y columnas en las variables X y Y.

```python
def Adquirir_imagen(image):
    img = cv2.imread(str(image)+'.jpg')
    #Acomodar el tamaño
    height, width = img.shape[:2]
    if (height or width) >= 256:
        X=512/height
        Y=512/width 
        img = cv2.resize( img, (int(Y*width), int(X*height)) )
    else:
        pass
    return img
```

# Función Imprimir Imagen

<ol>
  <li>**nombre_img**: Nombre de la imágen, debe ser string</li>
  <li>**img_imprimir**: Imágen que se quiere imprimir</li>
</ol>

```python
def Imprimir(nombre_img,img_imprimir,figure=None):
    fig = plt.figure(figure)
    ax = fig.add_subplot(1, 1, 1)
    ax.imshow(img_imprimir)
    plt.axis("off")
    plt.title(str(nombre_img))
    plt.show()
```

# Función para detectar Bordes

Se trabajó en el espacio HLS para manipular la luminancia y la saturación (Donde se obtuvo mejores resultados), luego se pasan estos canales através del filtro canny y se dibujan los bordes. La función recibe la imagén original y devuelve una máscara de la imágen con los bordes.

```python
def Contorno(image):
    img2 = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
    Imprimir("hls",img2)
    h,l,s= cv2.split(img2)
    ret, l2 = cv2.threshold(l,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
    edged = cv2.Canny(l2, 60, 150,apertureSize=3)
    Imprimir("edeged",edged)
    image1, contours, hierarchy = cv2.findContours(edged,cv2.RETR_TREE,
                                                   cv2.CHAIN_APPROX_SIMPLE)
    img = cv2.drawContours(image, contours, -1, (255,0,0), 2)
    
    return img
```

# Función Entropía

La entropía es "the function cost" del algoritmo. Recibe la imágen en forma de numpy array y calcula el histograma de la imágen en color. Luego de esto se transforma la matriz en un vector 1D para posteriormente aplicar la definición de entropía de shannon.

```python
def Entropia(imagen):
    histograma = cv2.calcHist([imagen],[0],None,[256],[0,256])
    histograma = histograma.ravel()/histograma.sum()
    AproxLogs = np.log2(histograma+0.00001)
    entropia = -1 * (histograma*AproxLogs).sum()
    return entropia
```

# Ubicar donde aplicar función Entropia

Cuando se está separando en super píxeles (Clusters), cada uno de ellos se almacena en una máscara binaria. Por ello, si se aplica la entropía directamente a esa imágen se vería alterada por los valores (0,0,0) negros que no nos interesan y no hacen parte del super píxel. Por ello, se almacenan los valores de los super píxeles en un numpy array y se aplica la Entropía sobre dicho arreglo. La función recibe la máscara de la imágen (o super píxel) y devuelve el arreglo para calcular la entropía.

```python
def Entropia_Mask(image_mask):
    array=[]
    for i in range(image_mask.shape[0]):
        for k in range(image_mask.shape[1]):
            x=image_mask[i,k] 
            if (x[0]==0 and x[1]==0 and x[2]==0):
                pass
            else:
                array.append(image_mask[i,k])
    array2=np.asarray(array)
    return array2
```

# Características de la imagen a segmentar

La siguiente clase describe el objeto: Imágen a ser segmentada:
<ol>
  <li>**self.image**: contiene la imágen original </li>
  <li>**self.k_cluster**: contiene el arreglo con los clusters de la imágen </li>
  <li>**self.cluster**: contiene la información individual de cada cluster: Entropía </li>
</ol>
De igual forma se construyen las funciones para cambiar y obtener informaciones del objeto:
<ol>
  <li>**getSegmentedImage**: regresa la imágen analizada con sus super píxeles</li>
  <li>**setSegmentedImage**: actualiza la imágen con los nuevos clusters </li>
  <li>**getKCluster**: regresa el arreglo con los clusters </li>
  <li>**setKCluster**: Dado que el algoritmo es recursivo, esta  función  actualiza los k clusters cada vez que se separa un super píxel </li>
  <li>**getEntropy**: Regresa la entropia de un determinado cluster </li>
  <li>**setEntropy**: Modifica o crea el valor de la entropía para un determinado cluster </li>
  <li>**getMaxKCluster**: Regresa el numero de clusters de la imágen </li>
</ol>

```python
class SegmentedImage:
    def __init__(self,image,cluster,k_cluster=None):
        self.image=image
        self.k_cluster=k_cluster 
        self.cluster=cluster
    def getSegmentedImage(self):
        return self.image
    def setSegmentedImage(self,original_image):
        self.image=mark_boundaries(original_image, self.k_cluster)
    def getKCluster(self):
        return self.k_cluster 
    def setKCluster(self,k_new_cluster,img_mask,old_kcluster,cluster_actual):
        k_cluster_old=np.unique(old_kcluster)
        b=1
        c=0
        for (i,segVal) in enumerate(np.unique(k_new_cluster)):
            if segVal==cluster_actual:
                for j in range(img_mask.shape[0]):
                    for k in range(img_mask.shape[1]):
                        x=img_mask[j,k] 
                        if (x[0]==0 and x[1]==0 and x[2]==0):
                            pass
                        else:
                            old_kcluster[j,k]=int(max(k_cluster_old))+b
                b+=1
            else:
                pass
        return old_kcluster    
    def getEntropy(self,Kcluster): #KCluster es el cluster o segmento actual
        return self.cluster[str(Kcluster)]
    def setEntropy(self,Kcluster,entropia):
        self.cluster[str(Kcluster)]=entropia
    def getMaxKCluster(self):
        return max(self.k_cluster)
```

# Algoritmo para cálculo de Super Píxeles

**¿Cómo funciona?**: el algoritmo separa inicialmente la imágen en k cluster iniciales y calcula la entropía de cada uno de ellos. Dado que la entropía mide la información de los colores em la imágen, mientras más pequeño sea, más probabilidades de pertenecer al mismo espacio de super píxel. Si la probabilidad de no pertencer es alta, ese super píxel se subdivide de nuevo en k super pixeles y se analiza de nuevo. Lo anterior se repite hasta conseguir los k clusters que cumplan con el criterio (que sea mayor al 25% de la entropía de la imágen original).

- La función recibe: 
    - **img**: imágen a analizar
    - **control_profundidad**: para determinar cuantos ciclos recursivos puede tener el algoritmo
    - **var_control**: para saber si nos encontramos en el primer ciclo de la recursión
    - **img_seg**: objeto (Imagen a segmentar)
- La función devuelve:
    El objeto (Imágen segmentada)


```python
def SuperPixel(img,control_profundidad=0,var_control=0,img_seg=None):
    #Necesitamos la imagen en float, so:
    image = img.astype('float')
    condition=Entropia(img)*0.25
    #Inicializar imagen segmentada
    numSegments=5
    if var_control==0:
        cluster_ini={}  
        img_seg=SegmentedImage(img,cluster_ini)
        img_seg.k_cluster=slic(image, n_segments = numSegments, sigma = 5)
        for (i,segVal) in enumerate(np.unique(img_seg.k_cluster)):
            mask = np.zeros(img.shape[:2], dtype = "uint8")
            mask[img_seg.k_cluster == segVal] = 255
            img_mask=cv2.bitwise_and(img_seg.image, img_seg.image, mask = mask)
            array_entropia=Entropia_Mask(img_mask)
            if array_entropia.size==0:
                img_seg.setEntropy(segVal,0)
            else:
                img_seg.setEntropy(segVal,Entropia(array_entropia))
            if img_seg.cluster[str(segVal)] > condition:
                control_profundidad += 1
                img_clustered=SuperPixel(img_mask,control_profundidad,1,img_seg)
            else:
                pass
    else:
        create_seg=slic(img, n_segments = numSegments, sigma = 5)  
        for (i,segVal) in enumerate(np.unique(create_seg)):
            mask = np.zeros(img.shape[:2], dtype = "uint8")
            mask[create_seg == segVal] = 255
            img_mask=cv2.bitwise_and(img, img, mask = mask)
            array_entropia=Entropia_Mask(img_mask)
            if array_entropia.size==0:
                img_seg.setEntropy(segVal,0)
            else:
                img_seg.setEntropy(segVal,Entropia(array_entropia))
            if img_seg.cluster[str(segVal)] > condition:
                img_seg.k_cluster=img_seg.setKCluster(create_seg,img_mask,
                                                      img_seg.k_cluster,segVal)
                if control_profundidad>3:
                    continue
                control_profundidad += 1
                img_clustered=SuperPixel(img_mask,control_profundidad,1,img_seg)
            else:
                pass
    return img_seg
```
